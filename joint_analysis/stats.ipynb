```python
umi_rna=pd.DataFrame()
for dataset in data_config:
    df=load_data(data_config[dataset]['rna_path'])
    umi=df.sum(axis=0)
    umi=pd.DataFrame(umi, columns=['n_umi'])
    umi['sample']=dataset
    umi_rna=umi_rna.append(umi)
umi_rna.to_csv('rna_umi.txt',sep='\t')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="sample",y="n_umi",data=umi_rna,palette="Set1",linewidth=1,fliersize=1)
plt.ylim(0,10000)
```

```python
gene_rna=pd.DataFrame()
for dataset in data_config:
    df=load_data(data_config[dataset]['rna_path'])
    genes=df[df>0].count(axis=0)
    genes=pd.DataFrame(genes, columns=['n_genes'])
    genes['sample']=dataset
    gene_rna=gene_rna.append(genes)
gene_rna.to_csv('rna_genes.txt',sep='\t')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="sample",y="n_genes",data=gene_rna,palette="Set1",linewidth=1,fliersize=1)
plt.ylim(0,10000)
```
